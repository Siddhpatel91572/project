```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Sample dataset
data = {
    "Screen_Time": [2, 3, 4, 5, 6, 7, 8],
    "CGPA": [8.6, 8.2, 7.8, 7.2, 6.8, 6.2, 5.9]
}

df = pd.DataFrame(data)

# Display dataset
print("Dataset:")
print(df)

# Correlation test
correlation, p_value = pearsonr(df["Screen_Time"], df["CGPA"])

print("\nCorrelation Coefficient:", correlation)
print("P-value:", p_value)

# Hypothesis decision
alpha = 0.05

if p_value < alpha:
    print("\nResult: Reject the Null Hypothesis")
    print("Conclusion: Smartphone usage significantly affects academic performance.")
else:
    print("\nResult: Fail to Reject the Null Hypothesis")
    print("Conclusion: No significant relationship found.")

# Visualization
plt.scatter(df["Screen_Time"], df["CGPA"])
plt.xlabel("Daily Smartphone Usage (Hours)")
plt.ylabel("CGPA")
plt.title("Smartphone Usage vs Academic Performance")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load dataset
df = pd.read_csv("/content/Students Social Media Addiction.csv") # kaggle dataset
# Display first few rows
print("Dataset Preview:")
print(df.head())

# Select required columns
x = df["Addicted_Score"]
# Convert 'Affects_Academic_Performance' to numerical (0 for No, 1 for Yes)
y = df["Affects_Academic_Performance"].map({'No': 0, 'Yes': 1})

# Correlation test
correlation, p_value = pearsonr(x, y)

print("\nCorrelation Coefficient:", correlation)
print("P-value:", p_value)

# Hypothesis testing
alpha = 0.05

if p_value < alpha:
    print("\nResult: Reject Null Hypothesis")
    print("Conclusion: Social media addiction significantly affects academic performance.")
else:
    print("\nResult: Fail to Reject Null Hypothesis")
    print("Conclusion: No significant effect found.")

# Visualization
plt.scatter(x, y)
plt.xlabel("Social Media Addiction Level")
plt.ylabel("Academic Performance Affected (0=No, 1=Yes)")
plt.title("Social Media Addiction vs Academic Performance Affected")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": [f"Student_{i}" for i in range(1, 501)],
    "Enroll": [f"ENR{i:04d}" for i in range(1, 501)],
    "Total Marks": np.random.randint(150, 501, 500)
}

df = pd.DataFrame(data)

df.head()

df.tail()

df["Percentage"] = (df["Total Marks"] / 500) * 100

def grade_assign(p):
    if p >= 90:
        return "A"
    elif p >= 75:
        return "B"
    elif p >= 60:
        return "C"
    elif p >= 40:
        return "D"
    else:
        return "F"

df["Grade"] = df["Percentage"].apply(grade_assign)

df.head()

grade_counts = df["Grade"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Grade Distribution of Students")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# New dataset for Business Partners and Shareholding Percentage
data_partners = {
    "Partner_Name": [f"Partner_{i}" for i in range(1, 501)],
    "Partner_ID": [f"PID{i:04d}" for i in range(1, 501)],
    "Shareholding_Percentage": np.random.randint(1, 101, 500) # Percentage from 1 to 100
}

df_partners = pd.DataFrame(data_partners)

print("Business Partners Dataset Head:")
display(df_partners.head())

print("Business Partners Dataset Tail:")
display(df_partners.tail())

# For visualization, we can categorize shareholding into bins or directly plot distribution
# Let's create bins for demonstration, similar to grades
def share_category_assign(percentage):
    if percentage >= 75:
        return "Major Shareholder"
    elif percentage >= 50:
        return "Significant Shareholder"
    elif percentage >= 25:
        return "Minor Shareholder"
    else:
        return "Small Shareholder"

df_partners["Share_Category"] = df_partners["Shareholding_Percentage"].apply(share_category_assign)

print("Business Partners Dataset with Share Categories Head:")
display(df_partners.head())

share_counts = df_partners["Share_Category"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(share_counts, labels=share_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis.colors)
plt.title("Shareholding Percentage Distribution of Business Partners")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the Students Social Media Addiction dataset
file_path = '/content/extracted_data/Students Social Media Addiction.csv'
df_social_media_new_project = pd.read_csv(file_path)

print("Dataset for New Project Head:")
display(df_social_media_new_project.head())

# Select new columns for correlation analysis
x_new = df_social_media_new_project["Avg_Daily_Usage_Hours"]
y_new = df_social_media_new_project["Sleep_Hours_Per_Night"]

# Correlation test for new variables
correlation_new, p_value_new = pearsonr(x_new, y_new)

print("\nCorrelation Coefficient (Avg Daily Usage vs Sleep Hours):", correlation_new)
print("P-value (Avg Daily Usage vs Sleep Hours):", p_value_new)

# Hypothesis testing for new variables
alpha = 0.05

if p_value_new < alpha:
    print("\nResult: Reject Null Hypothesis (H₀)")
    print("Conclusion: Average daily smartphone usage significantly affects sleep hours.")
else:
    print("\nResult: Fail to Reject Null Hypothesis (H₀)")
    print("Conclusion: No significant relationship found between average daily usage and sleep hours.")

# Visualization for new variables
plt.figure(figsize=(8,6))
plt.scatter(x_new, y_new)
plt.xlabel("Average Daily Smartphone Usage (Hours)")
plt.ylabel("Sleep Hours Per Night")
plt.title("Average Daily Smartphone Usage vs Sleep Hours")
plt.grid(True)
plt.show()
```
